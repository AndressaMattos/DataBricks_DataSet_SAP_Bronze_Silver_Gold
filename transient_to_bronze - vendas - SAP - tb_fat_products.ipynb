```
import great_expectations as ge
from pyspark.sql import functions as F
```

Carregamento

```
tmp_parquet_path = "/tmp/transient/tb_fat_products_1"
df = spark.read.format("parquet").load(tmp_parquet_path)
```

```
display(df)
```

Verificações

```
df_ge = ge.dataset.SparkDFDataset(df)

#Verificar se a coluna 'PRODUCTID' não possui valores nulos
expectativa_nulos_product = df_ge.expect_column_values_to_not_be_null(column='PRODUCTID')

#Outra coluna PRODCATEGORYID para verificação
expectativa_nulos_product_category = df_ge.expect_column_values_to_not_be_null(column='PRODCATEGORYID')

#Validar as expectativas
resultado_validacao = df_ge.validate()

if resultado_validacao['success']:
    print('Arquivo Ok!')
else:
    print('Erro na validação!')
```

```
df = df.withColumn("dt_ingestion", F.date_format(F.current_date(), "dd-MM-yyyy"))
```

```
display(df)
```

Salvar no formato Delta

```
delta = "/tmp/bronze/tb_fat_products"
df.write.partitionBy("dt_ingestion").format("delta").mode("overwrite").save(delta)
```
